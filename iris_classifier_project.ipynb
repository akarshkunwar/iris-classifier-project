```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
```

```python
columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class labels']
df = pd.read_csv('iris.dataset.data', names = columns)
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df, hue = 'class labels')
```

```python
data = df.values
X = data[:,0:4]
Y = data[:,4]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

trying different models

```python
##1: Support Vector Classifier Model
model1_svc = SVC()
model1_svc.fit(X_train, y_train)
```

```python
prediction1 = model1_svc.predict(X_test)

print(accuracy_score(y_test, prediction1)*100,'%')
```

```python
##2: Logistic Regression Model
model2_lr = LogisticRegression()
model2_lr.fit(X_train, y_train)
```

```python
prediction2 = model2_lr.predict(X_test)

print(accuracy_score(y_test, prediction2)*100,'%')
```

```python
##3: Decision Tree Classifier Model
model3_dtc = DecisionTreeClassifier()
model3_dtc.fit(X_train, y_train)
```

```python
prediction3 = model3_dtc.predict(X_test)
print(accuracy_score(y_test, prediction3)*100,'%')
```

```python
##4: Linear Discriminant Analysis Model
model4_ldc = LinearDiscriminantAnalysis()
model4_ldc.fit(X_train,y_train)
```

```python
prediction4 = model4_ldc.predict(X_test)
print(accuracy_score(y_test, prediction4)*100,'%')
```

```python
print(classification_report(y_test, prediction4))
```

```python
X_input = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])
prediction = model4_ldc.predict(X_input)
print("prediction of species: {}".format(prediction))
```
